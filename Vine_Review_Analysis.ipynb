```python
import pandas as pd
```

```python
vine_df = pd.read_csv("vine_table.csv")
vine_df.head()
```

```python
# Step 1

high_total_votes_df = vine_df.loc[vine_df["total_votes"] >= 20, :]
high_total_votes_df.head()
```

```python
# Step 2
vote_ratio = (high_total_votes_df["helpful_votes"])/(high_total_votes_df["total_votes"])
more_helpful_df = high_total_votes_df.loc[vote_ratio > 0.5, :]
more_helpful_df.head()
```

```python
# Step 3
paid_users_df = more_helpful_df.loc[more_helpful_df["vine"] == "Y", :]
paid_users_df.head()
```

```python
# Step 4
unpaid_users_df = more_helpful_df.loc[more_helpful_df["vine"] == "N", :]
unpaid_users_df.head()
```

```python
# Step 5

total_reviews = vine_df.shape[0]
five_star_reviews = (vine_df.loc[vine_df["star_rating"] == 5.0, :]).shape[0]
paid_reviews = paid_users_df.shape[0]
unpaid_reviews = unpaid_users_df.shape[0]
five_star_paid = (paid_users_df.loc[paid_users_df["star_rating"] == 5.0, :]).shape[0]
five_star_unpaid = (unpaid_users_df.loc[unpaid_users_df["star_rating"] == 5.0, :]).shape[0]

five_star_percent_paid = five_star_paid/paid_reviews
five_star_percent_unpaid = five_star_unpaid/unpaid_reviews


print(f'Total Reviews: {total_reviews:,}\nFive Star Reviews: {five_star_reviews:,}\nFive Star Percent (Paid): {five_star_percent_paid:.2%}\nFive Star Percent (Unpaid): {five_star_percent_unpaid:.2%}\nVine Reviews: {paid_reviews}\nNon-Vine Reviews: {unpaid_reviews:,}')
print(f'Five Stars (Vine): {five_star_paid:,}\nFive Stars (Non-Vine): {five_star_unpaid:,}')
```
